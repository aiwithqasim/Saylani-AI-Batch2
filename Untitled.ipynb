```python
import numpy as np
import pandas as pd
col=[]

for i in range(26):
    col.append(chr(i+65))

for i1 in range(26):
    for j1 in range(26):
        a=chr(i1+65) + chr(j1+65)
        col.append(a)
        
for i2 in range(26):
    for j2 in range(26): 
        for k in range(26):
            b=chr(i2+65) + chr(j2+65)+chr(k+65)
            col.append(b)

data=np.zeros((10,16384))
d1=pd.DataFrame(data,columns=col[:16384])
d1
```
